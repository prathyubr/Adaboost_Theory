### 4. Model Building

```python
X = cancer_data_df.drop(labels=['target'],axis=1)
y = cancer_data_df[['target']]
```

```python
X.shape,y.shape
```

#### ======================================================================================================

### What is Imbalance dataset?

Imbalanced classification is the problem of classification when there is an **unequal distribution of classes** in the training dataset.

```python
y.value_counts()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(6,8))
plt.pie(x = y.value_counts(),
    explode=[0.02,0],
    labels=['Benign','Malignant'],
    autopct='%2f')
plt.show()
```

#### Observation: Breast cancer is an imbalanced dataset.

#### ======================================================================================================

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=12,shuffle=True,stratify=y)
#Stratify - makes a split so that the proportion of values in the sample produced will be the same as the proportion of values provided before splitting.
```

#### y_train distribution

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(6,8))
plt.pie(x = y_train.value_counts(),
    explode=[0.02,0],
    labels=['Benign','Malignant'],
    autopct='%2f')
plt.show()
```

##### Without Stratify  - The Proportion of Benign - 64%, Malignant - 36%
##### With Stratify       - The Proportion of Benign - 63%, Malignant - 37%.[same as the distribution before splitting]

#### y_test distribution

```python
plt.figure(figsize=(6,8))
plt.pie(x = y_test.value_counts(),
    explode=[0.02,0],
    labels=['Benign','Malignant'],
    autopct='%2f')
plt.show()
```

##### Without Stratify  - The Proportion of Benign - 58%, Malignant - 42%
##### With Stratify       - The Proportion of Benign - 63%, Malignant - 37%.[same as the distribution before splitting]

### 5. Model Training

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

dt_model  = DecisionTreeClassifier(max_depth=3,class_weight={0:1,1:2})
#Weights associated with classes in the form ``{class_label: weight}``
dt_model.fit(X_train,y_train)

adb_model = AdaBoostClassifier()
adb_model.fit(X_train,y_train)
```

### 6. Model Testing | 7. Model Evaluation

#### Test data

```python
y_pred_test = adb_model.predict(X_test)

from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,classification_report
print('Train Accuracy        : ',  round(accuracy_score(y_test,y_pred_test),5))
print('Confusion Matrix      :\n', confusion_matrix(y_test,y_pred_test))
print('Classification Report :\n', classification_report(y_test,y_pred_test))
```
